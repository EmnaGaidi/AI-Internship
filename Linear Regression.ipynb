```python
import numpy as np 
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import PolynomialFeatures
```

# Degr√© 1

```python
# make_regression nous permet de g√©n√©rer des donn√©es
np.random.seed(0)
x,y = make_regression(n_samples=100,n_features=1,noise = 10)
plt.scatter(x,y)
```

```python
model = SGDRegressor(max_iter = 1000,eta0=0.001)
model.fit(x,y)
```

```python
print('Coeff R2 =', model.score(x, y)) 
plt.scatter(x, y) 
plt.plot(x, model.predict(x), c='red', lw = 3)
```

# Degr√© 2

```python
np.random.seed(0) 
# cr√©ation du Dataset 
x, y = make_regression(n_samples=100, n_features=1, noise=10) 
y = y**2 # y ne varie plus lin√©airement selon x ! 
# y = y**2 : Cette ligne √©l√®ve au carr√© toutes les valeurs de la variable cible y.
#Cela a pour effet de rendre la relation entre x et y non lin√©aire, car y est maintenant fonction de x √† la puissance 2.
# On ajoute des variables polyn√¥miales dans notre dataset 
poly_features = PolynomialFeatures(degree=2, include_bias=False) 
x = poly_features.fit_transform(x) 
plt.scatter(x[:,0], y) 
x.shape # la dimension de x: 100 lignes et 2 colonnes
```

```python
# On entraine le modele comme avant ! rien ne change ! 
model = SGDRegressor(max_iter=1000, eta0=0.001) 
model.fit(x,y) 
print('Coeff R2 =', model.score(x, y)) 
plt.scatter(x[:,0], y, marker='o') 
plt.scatter(x[:,0], model.predict(x), c='red', marker='+')
```
