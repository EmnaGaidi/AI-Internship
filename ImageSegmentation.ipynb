```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
image = mpl.image.imread("dog.jpg")
```

```python
plt.imshow(image)
```

```python
image.shape
```

*transform the image from to dimentional pixels into one and we keep the 3 of RGB*

```python
X = image.reshape(-1,3)
X.shape
```

```python
kmeans = KMeans(n_clusters = 2, n_init=10)
# the more clusters I add the more details I get
kmeans.fit(X)
```

```python
segmented_img = kmeans.cluster_centers_[kmeans.labels_]
segmented_img = segmented_img.reshape(image.shape)
plt.imshow(segmented_img/255)
# we divide by 255 to get values between 0 and 1
```

```python
# in order to get the images with png format
# cv2 uses RGB and not BGR that's why we need to convert
import cv2
cv2.imwrite("image1.png",cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
cv2.imwrite("image2.png",cv2.cvtColor(segmented_img.astype("uint8"),cv2.COLOR_BGR2RGB))
```
